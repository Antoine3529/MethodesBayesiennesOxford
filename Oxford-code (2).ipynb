## Algorithme Gibbs avec Metropolis-Hastings

```python
def calcul_r0(mu,r,n,N):
    R0 = 0 
    for i in range(N):
        R0 += r[i]*np.log(expit(mu[i]))+(n[i]-r[i])*np.log(1-expit(mu[i]))
    return R0

def calcul_r1(mu,alpha,beta_1,beta_2,year,b,r,n,N):
    R1 = 0 
    for i in range(N):
        R1 += r[i]*np.log(expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i]))+(n[i]-r[i])*np.log(1-expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i]))
    return R1
           
```

```python
def gibbs_oxf(nchain,burnin,r_0,n_0,n_1,r_1,year,K, prop_sd,b, init,init_mu,init_b):
    chain = np.zeros((nchain + 1, 4)) ## Colonne:alpha ,beta_1, beta_2,sigma
    chain_b = np.zeros((nchain + 1, K))
    chain_mu = np.zeros((nchain + 1, K))
    chain[0,:] = init
    chain_b[0,:] = init_b
    chain_mu[0,:] = init_mu
    acc_rates = np.zeros(5)
    acc_ratesb = np.zeros(K)
    acc_ratesmu = np.zeros(K)
    
    alpha , beta_1, beta_2, sigma = chain[0,:]
    b = chain_b[0,:]
    mu = chain_mu[0,:]
    
    for iteration in range(nchain):
        
        # Mise à jour alpha
        prop = np.random.normal(alpha,prop_sd[0])
        
        top = -(prop**2/(2*1E6)) + calcul_r1(mu,prop,beta_1,beta_2,year,b,r_1,n_1,K) # tau = 1E-6 dans l'énoncé donc sigma = 1000
        
        bottom = -(alpha**2/(2*1E6)) + calcul_r1(mu,alpha,beta_1,beta_2,year,b,r_1,n_1,K)
        
        acc_prob= np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            alpha = prop
            acc_rates[0] += 1
            
        # Mise à jour Beta_1
        prop = np.random.normal(beta_1,prop_sd[1])
        
        top = -(prop**2/(2*1E6)) + calcul_r1(mu,alpha,prop,beta_2,year,b,r_1,n_1,K)
        
        bottom = -(beta_1**2/(2*1E6)) + calcul_r1(mu,alpha,beta_1,beta_2,year,b,r_1,n_1,K)
        
        acc_prob= np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            beta_1 = prop
            acc_rates[1] += 1
            
        # Mise à jour de Beta_2
        
        prop = np.random.normal(beta_2,prop_sd[2])
        
        top = -(prop**2/(2*1E6)) + calcul_r1(mu,alpha,beta_1,prop,year,b,r_1,n_1,K)
        
        bottom = -(beta_2**2/(2*1E6)) + calcul_r1(mu,alpha,beta_1,beta_2,year,b,r_1,n_1,K)
        
        acc_prob= np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            beta_2 = prop
            acc_rates[2] += 1
        
        for i in range(K):
            
            # Mise à jour de mu
            prop = np.random.normal(mu[i],prop_sd[3])
            
            top = -(prop**2 / (2*1E6)) + r_0[i]* np.log(expit(prop)) + (n_0[i]-r_0[i])*np.log((1-expit(prop)))+ r_1[i]* np.log(expit(prop+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i])) + (n_1[i]-r_1[i])*np.log((1-expit(prop+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i])))
            
            bottom = -(mu[i]**2 / (2*1E6)) + r_0[i]* np.log(expit(mu[i])) + (n_0[i]-r_0[i])*np.log((1-expit(mu[i])))+ r_1[i]* np.log(expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i])) + (n_1[i]-r_1[i])*np.log((1-expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i])))
            
            acc_prob=np.exp(top - bottom)
            if np.random.uniform() < acc_prob:
                mu[i] = prop
                acc_ratesmu[i] += 1
        
        
            # Mise à jour de b
            
            prop = np.random.normal(b[i],prop_sd[4])
            
            top = -(prop**2 / (2*sigma**2))+ r_1[i]* np.log(expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+prop)) + (n_1[i]-r_1[i])*np.log(1-expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+prop))
            
            bottom = -(b[i]**2 / (2*sigma**2))+r_1[i]*np.log(expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i])) + (n_1[i]-r_1[i])*np.log(1-expit(mu[i]+ alpha + beta_1 * year[i] + beta_2*(year[i]**2-22)+b[i]))
            acc_prob=np.exp(top - bottom)
            if np.random.uniform() < acc_prob:
                b[i] = prop
                acc_ratesb[i] += 1
        # Mise à jour de sigma
        a = 0.001 + K / 2
        b1 = 0.001 + 0.5 * (np.sum(b ** 2))
        tau=np.random.gamma(shape =a, scale =b1 )
        sigma = 1 / np.sqrt(tau)

        # Mise à jour des chaines
        chain[iteration+1,0] = alpha
        chain[iteration+1,1] = beta_1
        chain[iteration+1,2] = beta_2
        chain[iteration+1,3] = sigma
        chain_b[iteration+1,:] = b 
        chain_mu[iteration+1,:] = mu
        
    acc_rates = acc_rates / (1.0 * nchain)
    acc_ratesb = acc_ratesb / (1.0 * nchain)
    acc_ratesmu = acc_ratesmu / (1.0 * nchain)
    
    #Burnin
    chain = chain[burnin:,:]
    chain_b = chain_b[burnin:,:]
    chain_mu = chain_mu[burnin:,:]
    
    return chain,chain_b,chain_mu,acc_rates,acc_ratesb,acc_ratesmu
        
    
```

```python
nchain = 10000
prop_sd = [0.075, 0.02, 0.0035, 0.77, 0.25]
init = [alpha,beta_1,beta_2,sigma]
init_mu = mu
init_b = b

chain,chain_b,chain_mu,acc_rates,acc_ratesb,acc_ratesmu = gibbs_oxf(nchain,1000,r_0,n_0,n_1,r_1,year,K, prop_sd,b, init,init_mu,init_b)
```

Beta2 : ce paramètre représente l'effet du carré de l'année de naissance moins 22 sur le logarithme du risque de cancer chez l'enfant. Ce terme capture l'effet non linéaire de l'année de naissance sur le risque de cancer chez l'enfant.
